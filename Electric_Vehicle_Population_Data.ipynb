### 1 How many unique car manufacturers (Make) are present and top 10 EV manufacturers?

```python

unique_makes = data['Make'].nunique()

print(f"Number of unique car manufacturers: {unique_makes}")
```

```python
# Plot top 10 EV manufacturers
make_counts = data['Make'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=make_counts.head(10).index, y=make_counts.head(10).values, palette='viridis')

plt.xticks(rotation=45)
plt.xlabel("Car Manufacturer (Make)")
plt.ylabel("Number of EVs")
plt.title("Top 10 Electric Vehicle Manufacturers")
plt.show()
```

### 2. What is the relationship between postal codes and EV adoption rates?

```python
# Count EVs per Postal Code
ev_by_postal = data['Postal Code'].value_counts().reset_index()
ev_by_postal.columns = ['Postal Code', 'EV Count']

# Display the top 10 postal codes with the highest EV adoption
print(ev_by_postal.head(10))
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=ev_by_postal.head(10)['Postal Code'], y=ev_by_postal.head(10)['EV Count'], palette='coolwarm')

plt.xticks(rotation=45)
plt.xlabel("Postal Code")
plt.ylabel("Number of EVs")
plt.title("Top 10 Postal Codes with the Highest EV Adoption")
plt.show()
```

### 3. What is the average electric range for different vehicle models?

```python
avg_range_per_model = data.groupby('Model')['Electric Range'].mean().reset_index()

print(avg_range_per_model.sort_values(by='Electric Range', ascending=False).head(10))
```

```python
top_10_models = avg_range_per_model.sort_values(by='Electric Range', ascending=False).head(10)

# Bar Plot
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_models['Electric Range'], y=top_10_models['Model'], palette='coolwarm')

plt.xlabel("Average Electric Range (miles)")
plt.ylabel("Vehicle Model")
plt.title("Top 10 Vehicle Models with the Highest Average Electric Range")
plt.show()
```

### 4. Analyze the Overall Distribution of Electric Range

```python
# Histogram of Electric Range Distribution
plt.figure(figsize=(10, 5))
sns.histplot(data['Electric Range'], bins=20, kde=True, color='blue')

plt.xlabel("Electric Range (miles)")
plt.ylabel("Number of Vehicles")
plt.title("Distribution of Electric Ranges in the Dataset")
plt.show()
```

### 5. What is the most common Electric Vehicle Type in the dataset 

```python
# Count the number of occurrences for each EV type
ev_type_counts = data['Electric Vehicle Type'].value_counts()

print(ev_type_counts)
print(f"\nMost common EV type: {ev_type_counts.idxmax()} ({ev_type_counts.max()} vehicles)")
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=ev_type_counts.index, y=ev_type_counts.values, palette="coolwarm")

plt.xlabel("Electric Vehicle Type")
plt.ylabel("Number of Vehicles")
plt.title("Distribution of Electric Vehicle Types")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
ev_type_counts.plot.pie(autopct='%1.1f%%', cmap="coolwarm", startangle=140)
plt.title("Electric Vehicle Type Distribution")

plt.show()
```

### 6. What percentage of EVs are eligible for Clean Alternative Fuel Vehicle (CAFV) incentives?

```python
import pandas as pd

# Load dataset
# data = pd.read_csv("your_dataset.csv")  # Uncomment if loading from a file

# Count the total number of EVs
total_evs = len(data)

# Count the number of EVs eligible for CAFV incentives
eligible_evs = data[data['Clean Alternative Fuel Vehicle (CAFV) Eligibility'] == 'Eligible'].shape[0]

# Calculate the percentage
percentage_eligible = (eligible_evs / total_evs) * 100

print(f"Total EVs: {total_evs}")
print(f"Eligible EVs: {eligible_evs}")
print(f"Percentage of EVs eligible for CAFV incentives: {percentage_eligible:.2f}%")
```

```python
# Count values for Eligible vs. Not Eligible
cafv_counts = data['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].value_counts()

# Bar Plot
plt.figure(figsize=(15, 5))
sns.barplot(x=cafv_counts.index, y=cafv_counts.values, palette="coolwarm")

plt.xlabel("CAFV Eligibility")
plt.ylabel("Number of Vehicles")
plt.title("EVs Eligible for Clean Alternative Fuel Vehicle (CAFV) Incentives")
plt.show()
```

```python
plt.figure(figsize=(7, 7))
cafv_counts.plot.pie(autopct='%1.1f%%', cmap="coolwarm", startangle=140)
plt.title("Percentage of EVs Eligible for CAFV Incentives")
plt.ylabel("")  # Hide y-label
plt.show()
```
